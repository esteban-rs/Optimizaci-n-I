```python
# functions
# read arguments as arrays and return and array
def f1(x1:np.array, x2:np.array):
    return x1*x1 - x2*x2

def f2(x1:np.array, x2:np.array):
    return 2*x1*x2

# find intersections
# def findIntersection(contour1, contour2):
    # Get 2D point from contours
    # p1 = [(e[0], e[1]) for e in contour1.collections[0].get_paths()[0].vertices]
    # p2 = [(e[0], e[1]) for e in contour2.collections[0].get_paths()[0].vertices]

    # s1 = geometry.LineString(p1)
    # s2 = geometry.LineString(p2)

    # return s1.intersection(s2)
```

```python
# Intervals for x1,x2
x1 = np.linspace(-20, 20, 200)
x2 = np.linspace(-20, 20, 200)

# Build two-dimentional grids from x1, x2
X, Y = np.meshgrid(x1, x2)

# Set plot
plt.title('Contour Plot')
contours_f1 = plt.contour(X, Y, f1(X, Y), colors = 'black', levels = [12])
contours_f2 = plt.contour(X, Y, f2(X, Y), colors = 'blue', levels = [16])

# Text labels
plt.clabel(contours_f1, inline = True, fontsize = 8)
plt.clabel(contours_f2, inline = True, fontsize = 8)

# Find Intersections and add to plot
# a = findIntersection(contours_f1,contours_f2)
# plt.scatter(a.x, a.y,  color='r', s=50)
# plt.scatter(-a.x, -a.y,  color='r', s=50)

# Intersections found in pdf
# (-4,-2), (4,2)
plt.scatter(-4,-2,  color='r', s=50)
plt.scatter(4, 2,  color='r', s=50)

plt.grid(True)
plt.xlabel('x1')
plt.ylabel('x2')

# show everything
plt.show()
```
