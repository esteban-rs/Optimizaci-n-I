2. Obtenga el mínimo de las siguientes funciones usando los algoritmos previos con punto inicial $ x^0 $ dado abajo. Adicionalmente, corra o ejecute el algoritmo para puntos iniciales $ x^0 $ elegidos aleatoriamente. Grafique $ (k, f_k) $ y $ (k, ||g_k||) $ para cada función. 

```python
def get_graf(f_hist, g_hist) :    
    plt.plot(range(f_hist.shape[0]), f_hist) 
    # naming the x axis 
    plt.xlabel('k') 
    # naming the y axis 
    plt.ylabel('f_k') 

    # giving a title to my graph 
    plt.title('Iteration vs Fuction Value') 

    # function to show the plot 
    plt.show() 

    plt.plot(range(g_hist.shape[0]), g_hist) 
    # naming the x axis 
    plt.xlabel('k') 
    # naming the y axis 
    plt.ylabel('g_k') 

    # giving a title to my graph 
    plt.title('Iteration vs Gradient Norm') 

    # function to show the plot 
    plt.show() 
```

3. Aplique la implementación del punto 1. para obtener el mínimo de $ f(X) $ para $ \eta \sim \mathcal{N} (0, \sigma) $ y $ \lambda, \sigma > 0 $. Grafique $ (t_i, y_i) $ y $ (t_i, x_i^* (\lambda)) $ en la misma figura.

```python
def get_special_graf (x_k_next, params = {}) :
    n  = x_k_next.shape[0]
    lam = params['lambda']
    eta = params['eta']
    sigma = params['sigma']
    
    t = [2.0 * i / (n - 1.0) - 1.0 for i in range(n)]
    y = [t[i] * t[i] + eta for i in range(n)]


    plt.plot(t, y) 
    # naming the x axis 
    plt.xlabel('t_i') 
    # naming the y axis 
    
    plt.plot(t, x_k_next, '+') 
    
    plt.legend(["t[i] vs y[i]", "t[i] vs x[i]*"])


    plt.show() 
```
